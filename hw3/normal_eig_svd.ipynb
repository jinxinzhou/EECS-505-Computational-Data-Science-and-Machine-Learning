````julia
using LinearAlgebra:norm
function evec_error(qhat::Vector,q::Vector)
    
   @assert isapprox(norm(qhat,2),1.0) ["qhat is not of unit norm"]
   @assert isapprox(norm(q,2),1.0) [" is not of unit norm"]
 
   alpha = qhat'*q 
   error = norm(qhat - alpha'*q,2) 

   return error 
end
````

## Determinant of a matrix 

The [determinant](https://en.wikipedia.org/wiki/Determinant) of a square matrix denoted by $\det A$ of $\det(A)$ is a scalar valued function of the elements of the matrix. It encodes important [geometric information](https://en.wikipedia.org/wiki/Determinant#Geometric_meaning) about the elements of the matrix. 


Suppose 

$$ A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}.$$


Then the determinant of $A$ is given by

$$\det(A) = ad - bc.$$

See [here](https://en.wikipedia.org/wiki/Determinant#Definition) for explicit formulas for matrices of sizex $3 \times 3$ and greater.  The command `det(A)` computes the determinant of `A` in Julia.

## Characteristic polynomial

Let $A \in \mathbb{F}^{n \times n}$. The **characteristic polynomial** of $A$ is defined as the polynomial  defined by 

$$p_A(z) = \det(z\,I_n - A),$$

where $I_n$ denotes the $n \times n$ Identity matrix.

## Eigenvalues via the characteristic polynomial

The **eigenvalues** of a square matrix $A \in \mathbb{F}^{n \times n}$  are the $n$ values of $z$ that **satisfy the equation**

$$p_A(z) = 0,$$

where $p_A(z)$ is the characteristic polynomial of $A$. 

We can compute the eigenvalues using the `eigvals` command in the `LinearAlgebra` package, as illustrated below.

````julia
using LinearAlgebra:eigvals
A = [2 1; 3 4]
@show eigvals(A);
````

We can use the `eigen` command in the `LinearAlgebra` package  to compute the eigenvalues and eigenvectors of a square matrix as illustrated below. 

````julia
using LinearAlgebra:eigvals
λ, Q = eigen(A)
````

An $n \times n$ matrix $A$ has $n$ eigenvalues and $n$ eigenvectors which satisfy the relation

$$A q_i = \lambda_i q_i \qquad \textrm{for } i = 1, \ldots, n,$$

We verify this relation numerically next.   

````julia
@show isapprox(A*Q[:,1],λ[1]*Q[:,1])
@show isapprox(A*Q[:,2],λ[2]*Q[:,2]);
````

We verify this numerically next. 

````julia
isapprox(A*Q,Q*Diagonal(λ))
````

THe world eigenvalue comes from the root *eigen* (in German) which means characteristic. The eigenvalues characterize some aspects of a matrix but not everything about it in ways we illustrate next.


We first consider an example of a real matrix with complex eigenvalues  and hence complex eigenvectors.

````julia
A = [3 -2; 4 -1]
λ, Q = eigen(A)
@show isapprox(A*Q,Q*Diagonal(λ)) ;
@show λ;
@show Q;
````

Next we show an example of a non-zero matrix that has eigenvalues equal to zero -- clearly the eigenvalues do not characterize the matrix in this case. 
    

````julia
A = [0 1; 0 -0]
λ, Q = eigen(A)
@show isapprox(A*Q,Q*Diagonal(λ)) ;
@show λ;
@show Q;
````

[Nilpotent matrices](https://en.wikipedia.org/wiki/Nilpotent_matrix#Example_1) are a class of non-zero matrices which have eigenvalues equal to zero.
    
We will now consider special matrices for which the eigenvalues and eigenvectors are characteristic of the matrix, in a way that we shall describe shortly. 

# Normal vs non-normal matrices

A square matrix $A \in \mathbb{F}^{n \times n}$ is **[normal](https://en.wikipedia.org/wiki/Normal_matrix)** if and only if

$$ A^H A = A A^H.$$

We illustrate some examples of normal matrices. 

## Orthogonal or unitary matrices

Let $Q \in \mathbb{R}^{n \times n}$. Then $Q$ is an **orthogonal matrix** if 
    
$$Q^T Q = QQ^T = I_n.$$
    
Suppose $Q \in \mathbb{C}^{n \times n}$ and $Q$ satisfies
    
$$Q^H Q = QQ^H = I_n,$$
    
then we refer to $Q$ as a **unitary matrix**.


**Theorem**: Sums of orthogonal (or unitary) matrices are **not** orthogonal (or unitary)

The above example constitutes a counter-example.

**Fact**: Let $Q \in \mathbb{F}^{n \times n}$ be an orthogonal (or unitary) matrix. 


Prove that:

-- $ Q^H$ is an orthogonal (or unitary)  matrix

-- $ Q V$ is orthogonal (or unitary) if and only if $V$ is an orthogonal (or unitary) matrix. Consequently, that the same applies for $QV^H$.


**Consequence**: Products of orthogonal (or unitary) matrices yield orthogonal (or unitary) matrices.


$Q$ is an orthogonal(unitary) matrix. => $QQ^H =Q^HQ= I $, so $(Q^H)^HQ^H = QQ^H = I$ and $Q^H(Q^H)^H = Q^HQ = I$. So $Q^H$ is an othogonal(unitary) matrix.

If $Q^HV$ is orthogonal(or unitary) matrix => $(Q^HV)^H(Q^HV)=(V^HQ)(Q^HV)=V^H(QQ^H)V=V^H IV=V^HV=I$ and $(Q^HV)(Q^HV)^H=(Q^HV)(V^HQ)=I$  =>(VV^H)=I. 
so V has to be an orthogonal (or unitary) matrix.

````julia
θ = π/4
Q = [cos(θ) sin(θ); -sin(θ) cos(θ)]
p_QtQ = heatmap(Q'*Q, yflip=:true, title = L"Q^HQ", aspect_ratio =:equal)
p_QQt = heatmap(Q*Q',yflip=:true,  title = L"QQ^H", aspect_ratio =:equal)
plot(p_QtQ , p_QQt , layout = (1,2), size = (800, 400))
````

This is a quick way of computationally ascertaining whether a matrix is indeed orthogonal (or unitary).

## Symmetric or Hermitian matrices

Let $A \in \mathbb{F}^{n \times n}$. Then $A$ is a **symmetric matrix** if 
    
$$ A = A^T,$$
    
where $^T$ denotes the matrix transpose.
    
If, on the other hand, 
    
$$A = A^H,$$

then $A$ is labeled  as a **[Hermitian matrix](https://en.wikipedia.org/wiki/Hermitian_matrix)**. 

Thus, symmetric (or Hermitian) matrices constitute a large class of normal matrices. Recognizing a symmetric (or Hermitian) matrix in disguise allows us to immediately label it as normal, as we illustrate next. 

## Spectral theorem for normal matrices

The theorem below illustrates a central important fact of normal matrices.


**[Theorem](https://en.wikipedia.org/wiki/Spectral_theorem#Normal_matrices)**:

IF $A$ is a normal matrix then we can express

$$ A = Q \Lambda Q^H,$$

where $Q$ is an orthogonal (or unitary) matrix and $\Lambda = \textrm{diag}(\lambda_1, \ldots, \lambda_n)$ is a diagonal matrix of eigenvalues.

We compute the eigenvalues and eigenvectors of a matrix as before.

````julia
A = [3 -2;-2 -1]
λ, Q = eigen([3 -2;-2 -1])
````

We verify that the eigenvector matrices computed are indeed orthogonal.

````julia
p_QtQ = heatmap(Q'*Q, yflip=:true, title =L"Q^HQ", aspect_ratio =:equal)
p_QQt = heatmap(Q*Q',yflip=:true,  title = L"QQ^H", aspect_ratio =:equal)
plot(p_QtQ , p_QQt , layout = (1,2), size = (800, 400))
````

Finally, we verify the identity $A = Q\Lambda Q^H$ next. 

````julia
@show Q*Diagonal(λ)*Q'
@show isapprox(A ,Q*Diagonal(λ)*Q');
````

This is the sense in which the eigenvalues and the eigenvectors characterize a normal matrix! Eigenvalues and eigenvectors also characterize [diagonalizable](https://en.wikipedia.org/wiki/Diagonalizable_matrix) non-normal matrices but this is beyond our present discussion. 

**Consequence**: If $A$ is normal and $A = Q \Lambda Q^H$ then

$$ A Q[:,i] = \lambda_i Q[:,i].$$

We verify this numerically next. 

````julia
println("First with i = 1")
@show A*Q[:,1]
@show λ[1]*Q[:,1]
@show isapprox(A*Q[:,1],λ[1]*Q[:,1]);
println("")
println("Next with i = 2")
@show A*Q[:,2]
@show λ[1]*Q[:,2]
@show isapprox(A*Q[:,2],λ[2]*Q[:,2]);
````

## Eigenvalues of symmetric or Hermitian matrices are always real 

In the above example, the eigenvalues were real. It turns out this is not an accident but a consequence of the following fact.


**[Fact](https://en.wikipedia.org/wiki/Hermitian_matrix#Spectral_properties)**: Symmetric and Hermitian matrices always have  real eigenvalues.


We shall utilize this property whenever we encounter a symmetric or Hermitian matrix.

## Application: Geometry of $y = Ax$ via the  eigen-decomposition

Let $A \in \mathbb{R}^{n \times n}$ be a symmetric  matrix so that we can express $A$ via its eigen-decomposition as 
$A = Q \Lambda Q^T$, where $Q$ is an orthogonal matrix and $\Lambda$ is the diagonal matrix of real eigenvalues. 

Consider the linear transformation given by $ x \mapsto y = A x$. We would like to understand this transformation in terms of the eigenvectors and eigenvalues of $A$. 

### Step 1:  Understanding the transformation $ x \mapsto Q^T x$

We begin by the vector

$$ z = Q^T x,$$

is geometrically related to the vector $x$.

To that end, we shall consider vectors $x \in \mathbb{R}^2$ on the unit circle so that $||x||_2 = 1$. All $ 2 \times 2$ orthogonal matrices $Q$ can be parameterized by the angle $\theta \in [0,2\pi)$ so that   

$$ Q_{\theta} = \begin{bmatrix} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{bmatrix},$$

Thus, if $x = \begin{bmatrix} x_1 & x_2 \end{bmatrix}^T$ then we have that 

$$ z = Q_{\theta}^Tx = \begin{bmatrix} \cos(\theta) x_1 + \sin(\theta) x_2 \\ -\sin(\theta) x_1 + \cos(\theta) x_2 \end{bmatrix}.$$

In the  plot generated by the following code segment, we pick a value for $\theta$ and then display a vector $x$ on the unit circle -- we can alter the  vector $x$ by changing  `x_angle`. Consequently the vector $Ax$ also changes. 

````julia
θ_degrees = 45
θ = θ_degrees*π/180 ## angle in radians
Q = [cos(θ) -sin(θ); sin(θ) cos(θ)]
A = Matrix(Q')
println("Change the value of x_angle to vary x and also Ax")
eigshow(A)  ##  rotate the x vector to see how  x is transformed 
````

We now check to see if these observations persist when we alter  $\theta$.

````julia
@manipulate for θ_degrees = (30, 45, 60, 90)
    θ = θ_degrees*π/180 ## angle in radians
    Q = [cos(θ) -sin(θ); sin(θ) cos(θ)]
    A = Matrix(Q')
    eigshow(A)  ##  rotate the x vector to see how  x is transformed 
end
````

**Fact**::

Indeed! The transformation $x \mapsto Q_{\theta}^Tz$ when $Q_{\theta}$ is an orthogonal matrix produces a vector that is related to $x$ via a clockwise rotation of $\theta$ radians!

In higher dimensions, something analog happens even though the notion of clockwise and anti-clockwise does not make sense there.



Via experimentation we have stumbled upon another important fact that generalizes beyond the two dimensional setting above.

**Theorem:** Suppose $x \in \mathbb{F}^n$ and let $Q \in \mathbb{F}^{n \times n}$ be an orthogonal or unitary matrix. 
    
Then  
    
$$||x||_2 = ||Qx||_2 = ||Q^Hx||_2$$.
    

A consequence of this is that when $||x||_2 = 1$, then so will $||Q^H x||_2$, as we have just discovered.
    
**Exercise**: Prove this statement.

We begin by noting that if $Q$ is an orthogonal or unitary matrix then so is $Q^H$. Thus showing $||x||_2 = ||Qx||_2$ for any orthogonal or unitary matrix. suffices to prove that 

$$||x||_2 = ||Q^H x||_2.$$

If we prove that

$$||x||_2^2 = ||Qx||_2^2 = ||Q^H x||_2^2,$$

then we are done.

To that end, note that


$$||Qx||_2^2 = (Qx)^H (Qx) = (x^HQ^H)(Qx) = x^H(Q^HQ)x= x^HIx=x^Hx =||x||_2^2$$




### Step 2: Understanding the transformation $x \mapsto Q x$

````julia
@manipulate for θ_degrees = (30, 45, 60, 90)
    θ = θ_degrees*π/180 ## angle in radians
    Q = [cos(θ) -sin(θ); sin(θ) cos(θ)]
    A = Matrix(Q)
    eigshow(A)  ##  rotate the x vector to see how  x is transformed 
end
````

The fact that when $x$ is changed along the unit circle then so will $Q^Tx$ is no surprise and is a consequence of the fact that 

$$||Q^Tx||_2 = ||x||_2 = ||Qx||_2.$$

That $Q_{\theta}x$ induces a **counter-clockwise** rotation in $x$ can be deduced from our observation that $Q_{\theta} x$ induces a clock-wise rotation in $x$. This is because when $Q$ is orthogonal, we have that $QQ^T x = I x = x$.


Thus 

$$x = I x = (QQ^T) x = Q (Q^Tx).$$

Thus if $Q^Tx$ induces a clock-wise rotation then $Q (Q^Tx) = Qw$ where $w = Q^Tx$ must rotate $w$ counter-clockwise to undo that rotation!


### Step 3: Understanding the transformation  $ x \mapsto \Lambda x$

````julia
Λ = [2 0; 0 1]
eigshow(Λ) ##  rotate the x vector to see how  x is transformed 
````

````julia
@manipulate for λ₁ = (2, 1), λ₂ = (1 , 2)
    Λ = [λ₁ 0; 0 λ₂]
    eigshow(Λ) ##  rotate the x vector to see how x is transformed 
end
````

That the transformations $x \mapsto \Lambda x$ transforms unit-norm vectors $x$ to vectors on an  ellipse with whose major and minor axes are governed by $\max(\lambda_1, \lambda_2)$ and $\min(\lambda_1,\lambda_2)$ can be explained mathematically, as follows.

Consider the vector

$$ y = \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} = \Lambda x = \begin{bmatrix} \lambda_1 & 0 \\ 0 & \lambda_2 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} \lambda_1 x_1 \\ \lambda_2 x_2 \end{bmatrix}.$$

This implies that $y_1 = \lambda_1 \, x_1$ and $y_2 = \lambda_2 \, x_2$. Since $||x||_2 = 1$, we have that

$$x_1^2 + x_2^2 = 1 \Rightarrow \left(\dfrac{y_1}{\lambda_1}\right)^2 + \left(\dfrac{y_2}{\lambda_2}\right)^2 = 1,$$

which is the equation of the ellipse for the transformed vector $\Lambda x$ that explaisn all we have observed from `eigshow(Λ)`.

### Step 4: Understanding the transformation $ x \mapsto Q \Lambda Q^T x$

We now consider the transformation $x \mapsto y = Ax$ in the setting where $A = Q \Lambda Q^Tx$.



We can view the transformation  $x \mapsto Q \Lambda Q^Tx$ as three transformations in series, each of which relate to the transformations from above.

-- Transformation 1: $x \mapsto w = Q^Tx$

-- Transformation 2: $ w \mapsto z = \Lambda w$


-- Transformation 3: $ z \mapsto y = Q z$.



**Theoretical Prediction**:

# The Singular Value Decomposition (SVD)

Let $A \in \mathbb{F}^{m \times n}$. Then the **full [singular value decomposition](https://en.wikipedia.org/wiki/Singular_value_decomposition)** of $A$ yields the relationship

$$ A = U \Sigma V^H,$$

where $U \in  \mathbb{F}^{m \times m}$ and $V \in \mathbb{F}^{n \times n}$ are orthogonal or unitary matrices and $\Sigma $ is an $m \times n$ diagonal matrix  with strictly non-negative entries.


**Definitions**:

-- the matrix $U$ is referred to as the matrix of **left** singular vectors

-- the matrix $V$ is referred to as the matrix of **right** singular vectors

--- the strictly positive entries of  $\Sigma$ are referred to as the **singular values**.

-- the number of singular values equals the  so-called **rank** of the matrix $A$. 

We can compute the full SVD  using the `svd` command in the `LinearAlgebra` package as:

```julia

U, σ, V = svd(A,full=:true)
```

**Important fact**: The `svd` command will return $\min(m,n)$ values in the `σ` variable and the elements of `σ` will be sorted in   **descending** order. 

Indeed, as we illustrate next. 

````julia
A = [1 2; 3 4; 5 6]
U, σ, V = svd(A,full=:true)
@show size(U)
@show size(V)
@show length(σ);
@show σ; 
````

The `U` and `V` matrices returned by the `svd` are orthogonal. We verify this using the `heatmap` command next.  

````julia
pU = plot(
    heatmap(U'*U, yflip =:true, aspect_ratio=:equal, title = L"U^HU", showaxis=:false),
    heatmap(U*U', yflip =:true, aspect_ratio=:equal, title = L"UU^H", showaxis=:false), layout = (1,2))
        
pV = plot(
    heatmap(V'*V, yflip =:true, aspect_ratio=:equal, title = L"V^HV", showaxis =:false),
    heatmap(V*V', yflip =:true, aspect_ratio=:equal, title = L"VV^H", showaxis =:false), layout = (1,2))        

plot(pU,pV, layout = (2,1), size = (900,600))
````

## Sum of outer-product representations of $A$ via its SVD

Let $A = U \Sigma V^H$ be the full SVD of $A$.

Then, we can express $A$ as 


$$A = \begin{bmatrix} U_1 & U_2 \end{bmatrix} \begin{bmatrix} \Sigma_{r \times r} & 0_{r,n-r} \\ 0_{m-r,r} & 0{m-r,n-r} 
    \end{bmatrix} \begin{bmatrix} V_1 & V_2 \end{bmatrix}^H = U_1 \Sigma_{r \times r} V_1^H,$$

where $U_1$ and $U_2$ are $m \times r$ and $m \times (m-r)$ sub-blocks of $U$ and $V_1$ and $V_2$ are $n \times r$ and $n \times (n-r)$ sub blocks of $V$, respectively.
Here $\Sigma_{r \times r}$ is an $r \times r$ diagonal matrix of with $\Sigma_{i,i}=  \sigma_i$ and $\sigma_1 \geq \ldots \geq \sigma_r$, by convention.

Expanding out these terms we can show that

$$ A = \sum_{i = 1}^{r} \sigma_i u_i v_i^H.$$

**Exercise**: Verify this identify.

Note that

$$ A= U_1 \Sigma_{r \times r} V_1^H = \begin{bmatrix} u_1 & \ldots & u_r \end{bmatrix} \begin{bmatrix} \sigma_1 & \ldots & 0 \\ \vdots & \ddots & \vdots \\ 0 & \ldots & \sigma_r \end{bmatrix} \begin{bmatrix} v_1^H \\ \vdots & v_r^H \end{bmatrix}.$$


FIrst consider the term

$$U_1 \Sigma{r \times r} = \begin{bmatrix} u_1 & \ldots & u_r \end{bmatrix} \begin{bmatrix} \sigma_1 & \ldots & 0 \\ \vdots & \ddots & \vdots \\ 0 & \ldots & \sigma_r \end{bmatrix} = \begin{bmatrix} u_1\sigma_1 & \ldots & u_r\sigma_r \end{bmatrix},$$


so that

$$ A = \left( \begin{bmatrix} u_1 & \ldots & u_r \end{bmatrix} \begin{bmatrix} \sigma_1 & \ldots & 0 \\ \vdots & \ddots & \vdots \\ 0 & \ldots & \sigma_r \end{bmatrix} \right) \begin{bmatrix} v_1^H & \vdots & v_r^H \end{bmatrix} = \sum_{i=1}^r \sigma_i u_i v_i^H.$$

This completes the proof.

The command `rank(A)` in the `LinearAlgebra` package returns the rank of a matrix. We can use this to verify the theoreticaly computational as below. 

````julia
m, n = 3, 4
A = rand(m,n)
@show r = rank(A)
U, σ, V = svd(A,full=:true)
@show isapprox(A,U[:,1:r]*Diagonal(σ[1:r])*V[:,1:r]');
````

## Ambiguities in the computed singular value triplets

Another important consequence of the sum of outer-products represenation of a matrix in terms of its singular values and vectors is the recognition that to **exactly reconstruct** the matrix $A$ from its singular values and singular vectors, we need the $r$ triplets $\{\sigma_i,u_i,v_i\}$ for $i = 1, \ldots, r$.


The phase (or sign) ambiguity in the singular vectors is a manifestation of the same phenomenon as in the eigenvector setting.

The singular values are **necessarily positive** for it to be an SVD.  

## The reduced SVD

A by-product of the fact that **every** matrix $A$ can be expressed as the sum of outer-products of its $r$ singular vectors is the $m-r$ left singular vectors $u_{r+1}, \ldots, u_{m}$ and the $n-r$ right singular vectors $v_{r+1},\ldots, v_{n}$ **play no role** with respect to the recconstruction of $A$.

This motivates the **reduced** or economy decomposition of the SVD where instead of computing the $m \times m$ matrix $U$ and  the $n \times n$ matrix $V$, we only compute the $m \times r$ matrix $U_1$ and the $n \times r$ matrix $V_1$, via the command

```julia
U, s, V = svd(A, full=:false) 

```

In fact `full =:false` is the default option when invoking the `svd` in Julia.   

````julia
A = [1 2; 3 4; 5 6]
U₁, σ, V₁ = svd(A,full=:false)
@show size(U₁)
@show size(V₁)
@show length(σ)
@show σ;
````

````julia
pU = plot(
    heatmap(U₁'*U₁, yflip =:true, aspect_ratio=:equal, title = L"U_1^HU_1", showaxis=:false),
    heatmap(U₁*U₁', yflip =:true, aspect_ratio=:equal, title = L"U_1U_1^H", showaxis=:false), layout = (1,2))
        
pV = plot(
    heatmap(V₁'*V₁, yflip =:true, aspect_ratio=:equal, title = L"V_1^HV_1", showaxis =:false),
    heatmap(V₁*V₁', yflip =:true, aspect_ratio=:equal, title = L"V_1V_1^H", showaxis =:false), layout = (1,2))        

plot(pU,pV, layout = (2,1), size = (900,600))
````

We now verify that the reduced SVD allows us to reconstruct the original matrix.

### Julia v1.1 versus Math for the reduced SVD

A crucial differnce relative to the full SVD is the fact that singular vector matrices returned by the reduced SVD are non necessarily orthogonal

**Julia v1.1 versus Math for the reduced SVD**:

In `Julia v1.1` , the `svd(A, full=:false)` command in Julia 1.1 will **always** return  $\min(m,n)$ singular values regardless of the rank $r$.

## Truncated SVD


Recall that the matrix $A \in \mathbb{F}^{m \times n}$ can be expressed in terms of its SVD as

$$ A = \sum_{i=1}^{r}\sigma_i u_i v_i^H,$$

where $r$ is the rank of $A$.

The rank $k$ **truncated SVD** of the matrix $A$ is the matrix


$$A_{(k)} = \sum_{i=1}^{k} \sigma_{i} u_i v_i^H,$$

where $1  \leq k \leq r$. We will examine in future codexes why we might do this. 
    
We can use the `svds` command in the `Arpack` package to compute the requisite $k$ singular vectors instead of computing the $\min(m,n)$ singular vectors first. We illustrate this next. 
    



````julia
using Arpack
A = randn(3,4)
k = 1 ## number of singular value triplets to compute 
Uk, σk, Vk = svds(A; nsv=k)[1] ## nsv = k is where we specify how many triplets to compute 
Ak = Uk*Diagonal(σk)*Vk' 
@show norm(A-Ak,2)
@show isapprox(A,Ak)
````

 The matrix $A_{(k)}$ will not equal $A$ whenever $k < r$, as suggested by the sum of outer-products representation. 

## Relationship to eigen-decomposition

There is an intimate relationship between the SVD and the eigenvectors of the matrices $AA^H$ and $A^HA$.

Specifically, it can be shown that 

$$ AA^H = U \Sigma \Sigma^H U^H,$$

and

$$ A^HA = V \Sigma^H \Sigma V^H,$$

so that the eigenvectors of $AA^H$ are the left singular vectors of $A$ and the eigenvetors of $A^HA$ are the right singular vectors of $A$.

The shared eigenvalues  of $AA^H$ and $A^HA$ are the squared singular values of $A$.

See [here](https://en.wikipedia.org/wiki/Singular_value_decomposition#SVD_and_spectral_decomposition) and [here](https://ocw.mit.edu/courses/mathematics/18-065-matrix-methods-in-data-analysis-signal-processing-and-machine-learning-spring-2018/video-lectures/lecture-6-singular-value-decomposition-svd/)  for more details.
