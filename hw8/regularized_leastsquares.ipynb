# Beyond Ordinary Least Squares

The (ordinary) least squares problem given by

$$x_{\sf LS} = \arg \min _x ||Ax- b||_2,$$

has the minimum-norm solution

$$x_{\sf LS} = A^{+} b,$$

where $A^{+}$ is the pseudo-inverse of $A$. 

There are many situations in practice where we want a solution that is "better" than the ordinary least-squares solution. By definition, no other solution can be better than the least-squares solution in terms of minmizing the residual $||Ax-b||_2$. So, we motivate the notion of better with an example, next.

## The need for regularization

Consider a function 
    
 $$f(t) = \exp^{-0.5\,t} \sin\left(\frac{\pi}{4}t \right),$$
    
that we want to estimate from samples of data $T_1, \ldots, T_n$.  

We compute and display the function in the next cell. 

````julia
n = 100
T = range(0; stop=10, length=n)
f(t::Number) = exp(-0.5 * t) * sin(π * t / 4)
xtrue = f.(T)
scatter(T, xtrue; xlabel="t", ylabel="f(t)")
````

We are given $m$ measurements of this signal of the form

$$b = A \,x_{\sf true} + z,$$

where $A$ is an $m \times n$ matrix, $z$ is noise and  

$$ x_{\sf true} = \begin{bmatrix} f(T_1) \\ \vdots \\ f(T_n) \end{bmatrix}.$$

We compute $b$ via this model as in the next cell.

````julia
m = 400
A = randn(m, n)
z = 2 * randn(m)
b = A * xtrue + z;
````

We now compute te display the least-squares solution and the associated minimum least-squares residual error.

````julia
x_ls = pinv(A) * b
error_ls = norm(xtrue - x_ls)
scatter(T, x_ls; color=:black, label="OLS", xlabel="t", ylabel="f(t)")
scatter!(T, xtrue; label="ground truth", )
title!("LS error = $(round(error_ls; digits=4))")
````

In this case the underlying signal is smooth in a way that the least squares solution is not. This is why we go beyond ordinary least squares -- we will try to obtain a "better" solution that will generically have a larger residual error with respect to $||Ax-b||_2$ but for which the recovered solution is better in a sense that we will describe next.

# Regularized Least Squares

We begin by considering the *constrained least squares* problem of the form

\begin{equation}
\arg\min_x \|b- A x\|_2^2 \qquad \textrm{ subject to }  \|D x\|_2 \leq \tau.
\end{equation}

Here we modifying the ordinary least squares solution with the norm constraint. It can be shown that this is equivalent to the regularized least squares problem

\begin{equation}
x_{\delta} = \arg\min_x \|b- A x\|_2^2 + \delta^2 \|D x\|_2^2,
\end{equation}

where $\delta$ is the regularization parameter. Changing $\delta$ changes $x_{\delta}$ -- we will revisit the problem of how to select it. 


We can recast the regularized least squares problem as an ordinary least squares problem of the form 

\begin{equation}
 x_{\delta} = \arg \min_x ||\tilde{A} x - \tilde{b}\,||_2^2,
 \end{equation}

for appropriately defined $\tilde{A}$ and $\tilde{b}$ respectively. 



**Exercise**:

Determine $\tilde{A}$ and $\tilde{b}$ in terms of $A$, $D$ and $\delta$ and complete the  `regularized_ls_Ab` function below accordingly. 

````julia
using SparseArrays: spzeros
"""
    Atil, btil  = regularized_ls_Ab(A, b, D, delta)
               
Inputs:
* `A` is an m x n matrix
* `b` is an m x 1 vector
* `D` is a d x n matrix
* `delta` is a number

Output: Atil and btil such that `||Atil*x - btil||_2^2 = ||b - A*x||_2^2 + delta^2*||D*x||_2^2`
"""
function regularized_ls_Ab(A::AbstractArray, b::Vector, D::AbstractArray, delta::Number)
        Atil = vcat(A, delta .* D)
        btil = vcat(b, spzeros(size(D,1)))
        return Atil, btil
end
````

The solution of the regularized least squares problem is then conveniently given by


$$x_{\delta} = \tilde{A}^{+} \tilde{b}.$$




## Application: Improved Signal Recovery with  Derivative Norm Constraint

We now illustrate how we can improve the ordinary least squares solution we obtained earlier.

**Key idea**:

We want a smoother solution -- one that does not oscillate as much as the least squares solution we recovered.


We will cast this as a regularized least squares problem by selecting a solution that minimizes the least squares subject to the **additional constraint** that the Eucliean norm of the (time) derivative of the solution is small.


To that end, we first need to determine the matrix $D$ in the regularized least-squares formulation. 

## The first difference matrix in 1-D

To approximate the derivative of a function $f(x)$
that is sampled on the grid $x_1, \ldots, x_n$ where 

$$x_{i+1} = x_{i} + \Delta,$$

a typical  [finite difference](https://en.wikipedia.org/wiki/Finite_difference) approach is:

\begin{equation*}
\left. \partial_x{f(x)} \right|_{x= x_{i}}
\approx
\dfrac{f(x_{i+1}) - f(x_{i})}{\Delta}
.\end{equation*}

When the sample spacing
is $\Delta = 1$,
this approximation simplifies to

\begin{equation*}
f'(x_{i}) := \left. \partial{f(x)}{x} \right\rvert_{x= x_{i} }
\approx f(x_{i+1}) - f(x_{i})
.\end{equation*}

We can express this relation
for all $x_i$ samples
via the matrix-vector product

\begin{equation*}
\begin{bmatrix}
f'(x_1) \\
f'(x_2) \\
\vdots \\
f'(x_n) \\
\end{bmatrix}
\approx D_n
\begin{bmatrix}
f(x_1) \\
f(x_2) \\
\vdots \\
f(x_n) \\
\end{bmatrix},
\end{equation*}

where $D_n$ is the so-called **first difference matrix**
defined by

\begin{equation*}
%\label{eq:hp074:1}
D_n = \begin{bmatrix}
-1 &  1 &        &        &    \\
   & -1 &      1 &        &    \\
   &    & \ddots & \ddots &    \\
   &    &        &     -1 &  1 \\
 1 &    &        &        & -1
\end{bmatrix}.
\end{equation*}

Here we choose to set
$D_n[n,1] = 1$,

which corresponds to the (perhaps unexpected) approximation

$$f'(x_{n}) \approx f(x_{1}) - f(x_{n}).$$

````julia
using SparseArrays ## what package should we use so spdiagm works
"""
Syntax:       Dn = first_diffs_1d_matrix(n)
               
Inputs:       n is a positive integer
               
Outputs:      Dn is an n x n sparse circulant first differences matrix
"""
function first_diffs_1d_matrix(n::Integer)
    Dn = spdiagm(0 => -1*ones(n), 1 => ones(n-1))
    Dn[n, 1] = 1 # periodic boundary conditions
    return Dn
end
````

We now compute and display the regularized least squares solution for $\delta = 40$. 

````julia
D(n::Integer) = first_diffs_1d_matrix(n)
````

````julia
δ = 40
Atil, btil = regularized_ls_Ab(A, b, D(size(A, 2)), δ) 
x_reg_ls  = pinv(Matrix(Atil)) * btil
error_reg_ls = norm(xtrue - x_reg_ls)  

scatter(x_reg_ls; label="reg. LS")
scatter!(x_ls; color=:black, label="OLS")
scatter!(xtrue; color=:red, label="ground truth")
title!("LS error = $(round(error_ls; digits=4)), Reg. LS error = $(round(error_reg_ls; digits=4))")
````

We now examine how the quality of the solution changes as a function of $\delta$. 

````julia
x_ls = pinv(A) * b
error_ls = norm(xtrue - x_ls)

@manipulate for δ in (0, 10, 100, 400, 4000)
    Atil, btil = regularized_ls_Ab(A, b, D(size(A, 2)), δ) 
    x_reg_ls  = pinv(Matrix(Atil)) * btil
    error_reg_ls = norm(xtrue - x_reg_ls)  

    scatter(x_reg_ls; label="reg. LS")
    scatter!(x_ls; color=:black, label="OLS")
    scatter!(xtrue; color=:red, label="ground truth")
    title!("LS error = $(round(error_ls; digits=4)), Reg. LS error = $(round(error_reg_ls; digits=4))")
end
````

We explore this further by varying $\delta$ over a wider range, computing the regularized least squares solution as a function of $\delta$.

````julia
error_ls = norm(xtrue - x_ls)
Dn = D(size(A, 2))
norm_derivative_true = norm(Dn * xtrue)
error_reg_ls = []
norm_derivative_ls  = []

δ_list = 10 .^ range(-3; stop=3, length=100)

for δ in δ_list 
    Atil, btil = regularized_ls_Ab(A, b, Dn, δ) 
    x_reg_ls  = pinv(Matrix(Atil)) * btil
    
    error_reg_ls = push!(error_reg_ls, norm(xtrue - x_reg_ls))
    norm_derivative_ls = push!(norm_derivative_ls, norm(Dn * x_reg_ls))
end
````

 We plot the error relative to the ground truth and the norm of the derivative of the regularized least squares solution.

````julia
n = length(δ_list)
p1 = scatter(
    δ_list, error_reg_ls;
    xlabel="delta",
    ylabel="MSE", 
    label="Reg. LS",
    legend=:bottomright
)
plot!(δ_list, error_ls * ones(n); color=:red, label="OLS")

p2 = scatter(
    δ_list, norm_derivative_ls; 
    xlabel = "delta",
    ylabel = "Norm Derivative", 
    label = "Reg. LS"
)
plot!(δ_list, norm_derivative_true * ones(n); color=:black, label="Ground truth")
plot(p1, p2; layout = (2, 1))
````

This provides a prescription on how we might choose $\delta$ if we knew the norm of the derivative of the ground truth signal we are trying to estimate. That is also the tricky part and where we the subjectivity in data science and machine learning comes in -- we need to have a sense of what a "better solution" looks like so we can regularize the OLS solution to better approximate the solution! 

## Analytical insights on the regularization parameter choice 

We noticed that that  $x_{\delta}$ tends to a constant vector as $\delta \to \infty$. This is a situation where we have over-regularized the least-squares problem and we can analytically understand the structure of the solution in this setting. 

Note that  when $\delta \to \infty$ we are essentially solving the problem

$$ \arg \min_x ||Dx||^2_2,$$

instead of the problem

$$ \arg \min_x  ||Ax-b||_2^2 + \delta^2 ||Dx||^2_2.$$

Thus the $A$ matrix does not play a role and what emerges is a solution that depends on the regularization matrix $D$ and **not** on the measurements $A$ and $b$.

In our setting here, when $D$ is the first difference matrix in 1-D, then it can be shown that as $\delta \to \infty$, $x_{\delta} \to  c \bf{1}$, where $\bf{1}$ is the vector of all ones and $c \neq $ is some constant which is a vector in the nullspace of $D$. So, in that limit, as we have seen above, the minimum norm solution will tend to the vector of all zeros which makes the solution irrelevant in practice.     


# Application: Image inpainting with LS Total Variation Regularization

We now consider another application of regularization with a derivative constraint. We will illustrate this with an example next.

````julia
mask = rand(size(Xt)...) .< 0.7
img = Gray.(Xt)
mimg = img .* mask
````

Our goal is fill in the missing pixels by regularization.

## Step 1: Express problem in the form $y = Ax$

Suppose $y \in \mathbb{R}^{mn}$ is a vectorized image that has been corrupted by inserting zeros at (known) pixel locations. We can encode this missing pixel information in $A$ by making it an $mn \times mn$ diagonal matrix such that


$$y = A x,$$


where 

$$
A_{kk} = \begin{cases}
1 & \text{if } y_k \text{ is not corrupted} \\
0 & \text{if } y_k = 0.
\end{cases}
$$

## Step 3: Express as regularized least squares problem 

We will find a better solution by solving the regularized least squares problem 

$$\arg\min_x \|y-\mathbf{A}x\|_2^2 \textrm{ subject to }  \|\mathbf{C}x\|_2  \leq \tau,$$

where the matrix $C$ will be the first difference matrix in 2-D and $\|\mathbf{C}x\|_2  \leq \tau$ imposes a  constraint on the norm of the derivative of the solution.

As before we will convert it to a least-squares problem of the form 

$$\arg\min_x \|y-\mathbf{A}x\|_2^2 + \lambda^2 \|\mathbf{C}x\|_2^2,$$

and recast it into a usual least squares problem of the form


$$\arg \min_x ||\tilde{A} x - \tilde{b}||_2.$$


````julia
λ = 1e-10
Atil, btil =  regularized_ls_Ab(A, y, C, λ)
x_reg_ls = Atil \ Vector(btil)
@show norm(x_reg_ls - xt)
Gray.(reshape(x_reg_ls, size(Xt)))
````

# Iterative algorithms for ordinary and regularized least-squares 

Recall that the least squares problem

$$
x_{\sf ls} = \arg \min_{x} ||Ax - b||_{2}
,$$

has the solution

$$x_{\sf ls} = A^{+} b,$$

where $A^{+}$ denotes the pseudo-inverse. When the nullspace of $A$ is non-trivial then $x_{\sf ls}$
is the minimum norm solution.

**When $A$ is large**, it can be computationally prohibitive to compute the SVD of $A$
and then its pseudo-inverse before computing $x_{\sf ls}$. In such settings, **iterative algorithms are used in practice**. We now describe two such iterative algortihms. 

## The gradient descent algorithm 

The gradient descent algorithm is an iterative algorithm where the basic iteration (or update rule) given by

$$
x_{k+1} = x_{k} - \mu \,A^{H} \left(A x_{k} - b\right).$$

It is called  gradient descent because

$$
\nabla_x \frac{1}{2} ||Ax-b||_2^2 = - A^{H} (A x_{k} - b),
$$

and so the iterate is updated by the gradient times the learning rate $\mu$.

It can be shown that $x_k$ will minimize $||Ax-b||_2^{2}$ as iteration $k \rightarrow \infty$ provided

$$0 < \mu < \dfrac{2}{\sigma_1^2(A)},$$

where $\sigma_1(A)$ is the largest singular value of $A$.

Write a function called `lsgd` that implements the above least squares gradient descent algorithm.

````julia
using LinearAlgebra
"""
    x = lsgd(A, b, mu, x0, nIters)
              
Inputs:
* `A` is a m x n matrix
* `b` is a vector of length m
* `mu` is the step size to use, and must satisy 0 < mu < 2 / norm(A)^2 to guarantee convergence
* `x0` is the initial starting vector (of length n) to use
* `nIters` is the number of iterations to perform
              
Output: x is a vector of length n containing the approximate solution loss is a vector of the gradient descent loss
              
Description:  Performs gradient descent to solve the least squares

    min_x || b - A x ||_2
"""
function lsgd(A::AbstractArray, b::AbstractVector, mu::Number, x0::AbstractVector, nIters::Integer)
    loss = zeros(nIters)

    x = x0
    for idx in 1:nIters
        x = x - mu .* A' * (A * x -b)
        loss[idx] = norm(A * x - b)
    end
    return x, loss
end
````

## Nesterov's accelerated gradient descent  algorithm

We now describe an **accelerated** gradient descent method due to Nesterov for the least squares problem
    
$$
x_{\sf LS} = \arg \min_{x} \|A x - b\|_{2}
$$

that c**onverges provably faster** than the standard gradient descent method. The method consists of the following iteration

$$
x_{k+1} = z_{k+1} - \mu\, A^T \, \left(A z_{k+1} - b\right),
$$

where

$$z_{k+1} = x_k + \left(\dfrac{t_{k} - 1}{t_{k+1}}\right)
\left(x_k - x_{k-1}\right)$$

and
$$t_{k+1} = \dfrac{1 + \sqrt{1 + 4 t_k^2}}{2}.$$

The algorithm is initialized with $t_0 = 0$ and $x_{-1} = x_0$. 

This algorithm converges to $x_{\sf LS}$ when $0 < \mu \leq 1/ \sigma_1^2(A)$, where $\sigma_1(A)$ is the largest singular value of $A$.

**Exercise**:

Write a function called `lsngd` that implements the above least squares accelerated gradient descent algorithm.

````julia
using LinearAlgebra
"""
    x = lsngd(A, b, mu, x0, nIters)

Inputs:
* `A` is a m x n matrix
* `b` is a vector of length m
* `mu` is the step size to use, and must satisfy 0 < mu <= 1 / norm(A)^2 to guarantee convergence
* `x0` is the initial starting vector (of length n) to use
* `nIters` is the number of iterations to perform

Output: `x` is a vector of length n containing the approximate solution

Description:  Performs Nesterov-accelerated gradient descent
              to solve the least squares problem

    min_x || b - A x ||_2
"""
function lsngd(A::AbstractMatrix, b::AbstractVector, mu::Number, x0::AbstractVector, nIters::Integer)
    loss = zeros(nIters)
    # Nesterov-accelerated gradient descent
    t = 0
    xLast = x0
    x = x0
    for idx in 1:nIters
        # t update
        tLast = t
        t = (1 + sqrt(1 + 4 * tLast^2))/2

        # z update (momentum)
        z = x + (tLast-1)/t*(x - xLast)

        # x update
        xLast = x
        x = z - mu * (A' * (A * z - b))
        loss[idx] = norm(A * x - b)
    end
    return x, loss
end
````

## (optional) Peak signal-to-noise ratio (pSNR) improvement due to regularization    

In image reconstruction problems, a popular performance metric is **peak-signal-to-noise-ratio (pSNR)**. The pSNR measures the quality of an estimate $\hat{X}$ of a true image $X_{\mathrm{true}}$, and is defined by

$$
\text{pSNR}\left(\hat{X}, X_{\mathrm{true}}\right) = 10\log_{10}\left(\frac{\max_{ij}\left|X_{\mathrm{true}}\right|^2_{ij}}{\displaystyle\frac{1}{mn}\displaystyle\sum_{i,j=1}^{m,n} \left|\hat{X}_{ij} - \left(X_{\mathrm{true}}\right)_{ij}\right|^2}\right).
$$

The units of pSNR are decibels (dBs), and it can take values between $-\infty$ and $\infty$, where *pSNR = ∞* if and only if $\hat{X} = X_{\mathrm{true}}$.

````julia
# Compute pSNR
using Statistics
function compute_pSNR(Xhat, Xtrue)
    return 10 * log10(maximum(Xtrue)^2 / mean((abs.(Xhat - Xtrue)).^2))
end
````
