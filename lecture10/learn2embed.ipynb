# Embedding a Matrix of Coordinates in a Distance Matrix

We first load a dataset corresponding to a set of points in 2-D arranged in the shape of a "3".

```julia
"""
Function for convenient plotting of a digit, represented as an n-by-2 matrix.
"""
function plot_digit(X::Matrix, point_labels::Vector=string.(1:size(X, 1)); kwargs...)
    plot(
        X[:, 1], X[:, 2];
        marker=:circle, 
        aspect_ratio=:equal,
        kwargs...)
    annotate!(X[:, 1] .+ 1.5, X[:, 2], point_labels)
end

"""
Function for convenient plotting of a digit, represented as an n-by-2 matrix. Modifies previous plot.
"""
function plot_digit!(X::Matrix, point_labels::Vector=string.(1:size(X, 1)); kwargs...)
    plot!(
        X[:, 1], X[:, 2];
        marker=:circle, 
        aspect_ratio=:equal,
        kwargs...)
    annotate!(X[:, 1] .+ 1.5, X[:, 2], point_labels)
end
```

```julia
ll = split("a b c d e f g h i j")
d = JLD.load("misaligned_digit1.jld")
X = d["X"]
@show size(X)
plot_digit(X, ll; xlim=(10, 40))
```

We examine the matrix `X` next. 

```julia
X
```

We are given an $n \times d$ matrix $X$ for the form

\begin{equation}
X = \begin{bmatrix} x_1^T \\ \vdots \\ x_n^T  \end{bmatrix},
\end{equation}

where $x_1, \ldots, x_n$ are $d$ dimensional vectors whose elements correspond to the $d$ coordinates of each of the $n$ points.



From the matrix $X$, we can construct a **distance matrix** $D$ such that

\begin{equation}
D_{ij} = D_{ji} = d(x_i,x_{j}),
\end{equation}

where $d(x_i, x_j)$ is the distance between points $x_i$ and $x_j$, and $d(\cdot,\cdot)$ is a [distance function](https://en.wikipedia.org/wiki/Distance).

The **Euclidean distance** is expressed as

$$d(x_i,x_j) = ||x_i - x_j||_2,$$

where $||\cdot||_2$ is the Euclidean $\ell_2$ norm. A distance matrix $D$ generated with this distance function is called a *Euclidean* distance matrix. 

**Exercise**: 

Create a function called `coords2dist` that takes as its input the $n \times d$ matrix `X` and a `distance` function, and returns the $n \times n$ distance matrix `D`. 

```julia
using LinearAlgebra: norm

"""
    D = coords2dist(X, distance)

Input: `X` is an n x d matrix whose rows contains the d coordinates of the n objects 

Output:
* `D` is an n x n matrix such that D[i, j] is the distance from object i to object j
* `distance` is the function used to measure the distance between coordinates of objet i and objet j
*  default `distance`  is `vecnorm` which computes the usual Euclidean norm
"""
function coords2dist(X::Matrix, distance::Function=norm)
    n, d = size(X)
    D = [distance(X[i, :] - X[j, :]) for i in 1:(n), j in 1:(n)]
    return D
end
```

We now use this function to compute the Euclidean distance matrix of the points corresponding the digit we plotted a moment ago.  

```julia
D = coords2dist(X)
```

We now display the distance matrix as a heatmap.

```julia
plot(
    heatmap_labeled(D, ll),
    plot_digit(X, ll; xlim=(10, 40)); 
    layout=(1, 2),
    size=(700, 350)
)
```

# Recovering Coordinates from a  Distance Matrix  


Suppose we have the distance matrix, but not the original points. Can we recover the coordinates of the points from $D$? Before we do so, we would like to first understand the extent to which we can solve this problem. 

To that end, consider a translation of points corresponding to the code in the next cell.   

```julia
# shift by +5 in x dimension, and +10 in y dimension
 = [5 10]
X_shifted = X .+ 
plot_digit(X, ll; xlim=(10, 45))
plot_digit!(X_shifted, ll)
```

We compute the Euclidean distance matrix `Ddigit_shift` of the translated points and display the heatmap  of `Ddigit-Ddigit_shift`.  

```julia
D_shifted = coords2dist(X_shifted)
@show norm(D_shifted - D)
heatmap_labeled(D_shifted - D, ll; title="D_shifted - D")
```

This is unsurprising,because if $\mu$ is the translation vector (e.g. `[5 10]`), then we have that 

$$ ||(x_i + \mu) - (x_j + \mu) ||_2 = \lVert x_i - x_j + (\mu - \mu)\rVert_2 =  ||x_i - x_j||_2.$$


##  Translation of the coordinates does not change the distance matrix

This property holds for any distance function, and is called *translational invariance*. Intuitively, if you move two points by the same distance in the same direction, the distance between them does not change:

$$d(x_i + , x_j + ) = d(x_i, x_j).$$
            
Thus, **given a distance matrix**, Euclidean or otherwise, **we cannot uniquely recover the coordinates**. If $X$ is a set of coordinates that gives us a particular $D$, then $X .+ \mu$ (for *any* $\mu$) gives us the exact same distance matrix. 


## Scaling of the points scales the distance matrix 

Suppose we were to **scale** the points by $\alpha$. Then we have that

$$d(\alpha x_i,\alpha x_j) = \alpha d(x_i,x_j),$$

for any distance function.

This reveals that the distance matrix of $\alpha X$ is different by a factor of $\alpha$ from the distance matrix of $X$.
Thus, when $\alpha \neq 1$, scaling *changes* the distance matrix, regardless of the distance function used, in a way that translation does not.  

We now examine the effect of rotating the points (via a coordinate transformation) on the Euclidean distance matrix.  

```julia
 = /4
@show Q = [cos() -sin(); sin() cos()]
X_rotated = (X .+ )*Q
plot_digit(X, ll; xlim=(10, 70), ylim=(-10, 50))
plot_digit!(X_rotated, ll; marker=:square)
```

```julia
D_rotated = coords2dist(X_rotated)
@show norm(D_rotated - D)
heatmap_labeled(D_rotated - D, ll; title="D_rotated - D")
```

## Rotation of the points does not change the Euclidean distance matrix

Mathematically, if $Q$ is an orthogonal matrix, we would like to relate the distance matrix whose $(i,j)$-th element is 

$$||Q x_i - Q x_j||_2$$

to the distance matrix whose $(i,j)$-th element is $||x_i -x_j||_2$.

## What set of coordinates can we expect to recover from a Euclidean distance matrix?

The analysis reveals that if the matrix $X$ has at its rows, points whose coordinates (in the columns)  give us a particular $D$, then we will obtain the same $D$ from $(X .+ \mu) Q $ for *any* $\mu$ and any $d \times d$ orthogonal matrix $Q$. 
 
An important consequence is that if we are given a Euclidean distance matrix, then **we cannot expect to uniquely recover** a set of coordinates with the desired inter-point distances. 

Our analysis reveals that if we find one set of points that yields the given Euclidean distance matrix, then we can obtain another equivalent set of points via a translation and/or rotation.  This defines an equivalence class of recoverable points and our goal will be to receover a member of this class. 

## When does rotation of the points change the distance matrix?

The fact that the **Euclidean distance matrix** is  the same if we rotate the points is a conseqeuence of the orthogonal/unitary invariance of the Euclidean norm.

This is because  when $d(x_i,x_j)$ is the  Euclidean or $\ell_2$ norm of the difference then

$$d(Q x_i, Qx_j) = d(x_i, x_j),$$

for any orthogonal (or unitary) matrix $Q$.

If we used a distance function such that

$$d(Q x_i, Qx_j) \neq d(x_i, x_j),$$

for some orthogonal or unitary $Q$, then we would rotate the points and the **distance matrix would change**.


We illustrate this with an example next. Let 

$$d(x_i, x_j) = \parallel x_i - x_j \parallel_1,$$

where the $\ell_1$ norm of the $d$ dimensional vector $z$ is defined as

$$\parallel z \parallel_1 = \sum_{k=1}^{d} |z_k | .$$


We compute the corresponding $\ell_1$ distance matrix of the rotated digit and compare it to the $\ell_1$ distance matrix of the original digit in the next cell.  


```julia
l1_distance(x) = sum(abs, x)
D_l1 = coords2dist(X, l1_distance)
D_l1_rotated = coords2dist(X_rotated, l1_distance)
@show norm(D_l1_rotated - D)
heatmap_labeled(D_l1_rotated - D_l1, ll; title="D_l1_rotated - D_l1")
```

This is a quick computational "proof" that if we rotate a set of points, then the associated $\ell_1$ distance matrix changes as well. The $\ell_1$ norm is not unitary-invariant! We do not have to restrict our distances to be unitarily invariant but we will next describe an algorithm for recovering a set of coordinates that match a (noise-free) Euclidean distance matrix.
    


# Recovering coordinates from a Euclidean distance matrix

It can be shown that the **multidimensional scaling (MDS) coordinates** that match a given Euclidean $\mathbb{R}^{n \times n}$ distance matrix $D$ can be computed as follows. 


**Step 1**

Form the  matrix

\begin{equation}
K = -\dfrac{1}{2} P S P^T,
\end{equation}

where $S$ is the squared distance matrix 

$$ S = D.^2,$$ 

and 

$$ P = I - \frac{1}{n} 1 1^T.$$

**Step 2**:

Compute the eigen-decomposition of $K$ such that


$$K = U \Lambda U ^T.$$

**Step 3**:

Asssume that the embedding dimension of the coordinates $d$ is known (or given). The **MDS coordinates** of the $n$ points are given by the rows of the matrix

\begin{equation}
 X_r = U[:,1:d] \Lambda^{1/2}[1:d,1:d].
\end{equation}

**Practical consideration**:
 
Since we expect MDS to return real-valued coordinates, we will only take the values of $\Lambda$ that are non-negative in Step 3.



**Exercise**: 

Write a function called `dist2coords` that takes as its input the matrix $D$ and embedding dimension $d$, and returns the $n \times d$ matrix `Xr` whose rows are the MDS coordinates of the $n$ points.

Since the coordinates returned by MDS are unique only up to rotation and translation, we will impose additional constraints on `Xr` to make it unique. In particular, for this problem, you should transform `Xr` so that

+ The $i$th column of `Xr` is computed from the $i$th largest eigenpair of $K$ (reorder columns)  

**Note:** We use `sortperm()` to accomplish this below.

+ The actual $d$ used is capped at the number of positive eigenvalues of $K$ (positive eigenvalues only)

+ The rows of `Xr` have zero centroid (translate rows so that the average of all rows is a length-d vector whose elements are all zero)

+ The largest magnitude element of each column of `Xr` is positive (flip column signs)

The function `dist2coords` should also return the eigenvalues of the matrix $K$.

```julia
using LinearAlgebra: eigen, I, Diagonal
using Statistics: mean

"""
    Xr,  = dist2coords(D, d)

Inputs:
* `D` is an n x n matrix such that D[i, j] is the distance from object i to object j
* `d` is the desired embedding dimension. The value used should be 
capped at the number of positive eigenvalues of K

Outputs:
* `Xr` is an n x d matrix whose rows contains the relative coordinates
 of the n objects and e the n eigenvalues of K
* `` are the eigenvalues of K
 
Note: MDS is only unique up to rotation and translation, so we
enforce the following conventions on Xr:

* [ORDER] Xr[:, i] corresponds to ith largest eigenpair of K
* [EIG] Actual d used is min(d, # positive eigenvalues of K)
* [CENTER] The rows of Xr have zero centroid
* [SIGN] The largest magnitude element of Xr[:, i] is positive
"""
function dist2coords(D::Matrix, d::Integer)
    n = size(D, 1)

    # Compute correlation matrix
    S = D.^2  
    P = I - 1/n .* ones(n,1) .*ones(1,n) 
    K = -0.5 * P * S * P' 
    K = 0.5 * (K + K') # Force symmetry to eliminate any round-off errors

    # Compute relative coordinates
    , U = eigen(K)
    d = min(d, count( .> 0)) # Apply [EIG] 
    idx = sortperm(; rev=true)
    Xr = U[:, idx[1:d]] * Diagonal(sqrt.([idx[1:d]])) # Apply [ORDER] 

    # Apply [CENTER]
    Xr .-= mean(Xr; dims=1) ## TODO: Fill in ?? -- do we average over rows or columns?

    # Apply [SIGN]
    Xr .*= sign.(Xr[findmax(abs.(Xr); dims=1)[2]]) ## TODO: Fill in ??

    return Xr, 
end
```

## MDS coordinate recovery guarantee


**Theorem**:

If the Euclidean distance distance $D$ was generated from $n$ (unknown) points $x_1, \ldots, x_n \in \mathbb{R}^{d}$ such that 


$$D_{ij} = || x_i - x_j||_2,$$

then the MDS algorithm will recover points such that

$$|| X_r[:,i] - X_r[:,j]||_2 = D_{ij}.$$



## Example: Recovery of 2-D coordinates from a Euclidean Distance Matrix

We now use the MDS algorithm to compute and plot the coordinates from the distance matrix we considered earlier. We set `d = 2` when calling the `dist2coords` function because we know that the points are embedded in two dimensions. 

```julia
Xest, eigK = dist2coords(D, 2)
plot_digit(Xest, ll; label="MDS")
plot_digit!(X; marker=:square, label="Known", legend=:topleft)
```

We reconcile this seeming "failure" of the MDS algorithm next. 

Re-examining the MDS coordinate recovery guarantee theorem in this light reveals that the theorem makes no claim that
the MDS coordinates recovered in `Xr` will be related in any way to the true coordinates; the theorem merely states that `Xr` will yield the same distance matrix as the (unknown) true coordinates of the points!


Indeed! To that end, we now import the `myprocrustes` module. Then execute the following code cell to load it into the program scope.

## Algorithmically determining the embedding dimension 

In the previous example, we set `d = 2` in the `dist2coords` function because we knew the embedding dimension from the geometry of the points.
    
We now discuss how we might reason about the embedding dimension in practice if we are given the distance matrix and do not know anything what the points represent. 

To that end, we will discuss how the eigenvalues of the matrix $K$ convey this information. Recall that the matrix $K$ is related to  the distance matrix $D$ via $K = -0.5 PSP$. 

We begin by computing  the eigenvalues of $K$  and display them as in the next cell.

```julia
scatter(eigK; xlabel="index", ylabel="eig(K)")
```

Indeed. This is why we had to sort it explicilty from largest to smallest in the `dist2coords` function. And now we list the eigenvalues so we can more closely examine the numeric values taken by the smaller eigenvalues. 

```julia
println("Eigenvalues of K equal \n")
eigK
```

We determine the number of numerically significant positive eigenvalues of the matrix $K$  next.  

```julia
@show rankK = sum(eigK .> length(eigK)*eps(eigK[end]));
```

That the numerical rank of $K$ equals $2$ is no coincidence; that is the dimension of the underlying points from which $D$ was constructed in the first place. 

The rank of $X_r$ equals $d$ and this is why the eigenvalues of $K$ are often used to infer the embedding dimension.


**Determining $d$ algorithmically**:

When $d$ is unknown and

-- $D$ is noise-free then we set $d$ to equal the number of *numerically significant* positive eigenvalues of $K$

# Euclidean MDS embeddings for non-Euclidean distance matrices 

Euclidean MDS can be applied to generate **Euclidean embeddings** from distance matrices that are **not Euclidean**. We illustrate this next by applying MDS to the $\ell_1$ distance matrix of the digit.  

```julia
X_l1_est, eigK = dist2coords(D_l1, 2)
plot_digit(X_l1_est, ll; label="l1 MDS")
plot_digit!(X, ll; marker=:square, label="Known", legend=:topleft)
```

We use Procrustes to align the digits and display the aligned digit below along with a heatmap of the difference in the $\ell_2$ (or Euclidean) and $\ell_1$ distance matrices.  

```julia
X_l1_aligned, _ = procrustes(X, X_l1_est)

pdigit_l1 = plot_digit(X_l1_aligned, [""]; legend=:topright, label="l1 aligned MDS")
plot_digit!(X, ll; marker=:square, ylim=(10, 60), xlim=(0, 50), label="Known")

plot(
    heatmap_labeled(D - D_l1, ll; title="D_l2 - D_l1"),
    pdigit_l1,
    layout=(1, 2),
    size=(700, 350)
)
```

The reason why MDS reconstruction with the $\ell_1$ distance matrix is pretty good is because $D_{\ell_1} \approx D_{\ell_2}$ and so the MDS coordinates are approximately equal as well.


Mathematically, it can be shown that the worst case relative error in the MDS coordinates obtained from the $\ell_1$ distance matrix governed by the relative norm of the difference between the $\ell_2$ and $\ell_1$ distance matrices, which is given by

$$ ||D_{\ell_2} - D_{\ell_1}||_2/||D_{\ell_2}||_2.$$

We compute it for this example next. 

```julia
@show norm(D - D_l1)/norm(D);
```

The computation reveals that we can expect a 30% **worst-case** error. In this case, the worst-case is not realized and hence we get a better reconstruction than we might have feared.


This is not always the case though when we use MDS on non-Euclidean distance matrices. Appplying MDS to non-Euclidean distance matrices does often (but not always) work better than we might have feared. We illustrate this with additional examples next. 

## Example: City coordinates from the inter-city flight (or geodesic) distance matrix

We now show MDS can be applied to recover the coordinates from a *non-Euclidean* inter-city airline distance matrix. There are two **sources of deviation** from the ideal picture. These have to do with:

1) the cities are technically located in 3-D on the surface of the earth and not on a 2-D plane. So ideally we should get three outlying eigenvalues

When $d$ is unknown and

-- $D$ is noise-free then we set $d$ to equal the number of *numerically significant* positive eigenvalues of $K$

-- $D$ is noisy or has missing entries then $d$ is a free parameter that is equal to the number of *statistically significant* positive eigenvalues of $K$.


The eigen-spectrum of $K$ reveals information on the embedding dimension (via the number of "outlier" positive eigenvalues) and also via the error due to noise or discrepancy between how the distance is measured in the data and the Euclidean distance we are assumig implicitly when we solve for the MDS coordinates.

#  Clustering and unsupervised learning  

Clustering is the process of grouping similar entities together. **Clusters** in data refer to a collection, often a subset, of points that are similar or close  according to a similarity metric. Consider the set of points in the data below. 

```julia
X1 = randn(2, 400) 
X2 = randn(2, 400) 
X = hcat(X1 .+ [10; 5], X2 .+ [0; 5])
scatter_nd(X')
```

The goal of this unsupervised machine learning technique is to find similarities in the data point and group similar data points together. In unsupervised learning there are no labels and we use cluster analysis to automatically discover the labels. We describe one such algorithm for automatically discovering clusters.


## The k-means algorithm for clustering

The algorithms builds on the intuition you have described earlier. We define a cluster by the set of points that are closer to other points in the group than to points in a different group.

Suppose we wish to cluster the data consisting of points $x_1, \ldots, x_n$, organized as columns of a matrix $X$ into $k$ groups.  


The kmeans algorithm proceeds as follows:

**Step 1**:

Randomly assign the points to the $k$ clusters

**Step 2**:

For cluster $i = 1, \ldots, k$, we compute the *cluster centroid* $c_i$

**Step 3**:

Compute the distance $D[i,j]$ of each point $x_j$  to the cluster centroid $c_i$   for $j = 1, \ldots, n$


**Step 4**:

Recompute the cluster assignment of each point by determining which cluster centroid $x_i$ is closest to. In other words,  

$$\textrm{Cluster assignment of } x_j = \underset{i = 1, \ldots, k }{\arg\min} \,\, D[i,j] $$


Steps 2-4 are then repeated for a certain number of iterations. 

**Exercise**:

Implemnt the kmeans algorithm by completing the  `simple_kmeans` function below.

```julia
using Statistics: mean
using LinearAlgebra: norm

"""
Input: 
* X = d x n data matrix where n is number of samples and 
X[:,i] is a d-dimensional vector containing the d coordinates 
of the i-th sample 
* k = # clusters
* iters = # iterations to run the algorithm
* cluster_idx = initial assignment of data into clusters

Output: cluster_idx, C = matrix of Centroids corresponding to 
clusters and intra_cluster_sq_error 
"""
function simple_kmeans(
        X::Matrix, 
        k::Integer, 
        iters::Integer=100,
        cluster_idx::Vector=rand(1:k, size(X, 2))
    )
    
    # Initialize arrays
    C = zeros(size(X, 1), k)
    within_cluster_sqerror = zeros(k)

    for iter_idx in 1:iters
        D = zeros(k, size(X, 2))

        # Step 1: 
        # Computer centroid of each cluster and store in columns of C
        # Also compute squared distance of each point to its cluster's centroid
        for i in 1:k
            cluster_i_idxs = findall(cluster_idx .== i)
            cluster_i = X[:, cluster_i_idxs]
            C[:, i] = mean(cluster_i; dims=2)
            within_cluster_sqerror[i] = sum(abs2, cluster_i .- C[:, i])
        end
        
        # Step 2:
        # Compute distance from each point to cluster centroids 
        # D[i,j] = distance of point j to cluster centroid i
        for i in 1:k 
            for j in 1:size(X, 2)
                D[i, j] = norm(C[:, i] - X[:, j])
            end
        end
    
        # Step 3:
        # Compute new cluster assignment based on which centroid each point is closer to
        cluster_idx = mapslices(argmin, D; dims=1) |> vec
    end
    
    intra_cluster_sqerror = sum(within_cluster_sqerror)
    
    return cluster_idx, C, intra_cluster_sqerror  
end
```

```julia
X1 = randn(2, 400)
X2 = randn(2, 400)

using Random

@manipulate for iters in (0, 1, 10, 100, 1000),  in (1, 0.1, 0.01), k in (2, 1, 3)
    X = hcat(X1 .+  * [10; 5], X2 .+  * [0; 5])
    
    Random.seed!(1)
    cluster_idx = rand(1:k, size(X, 2))
    cluster_idx = simple_kmeans(X, k, iters, cluster_idx)[1][:]
    scatter_nd(X[:, cluster_idx .== 1]')
    scatter_nd!(X[:, cluster_idx .== 2]')
    scatter_nd!(X[:, cluster_idx .== 3]')
end
```

## Application: Finding and visualizing clusters in handwritten digits

We now illustrate how MDS can be used to display patterns within handwritten digits. 

```julia
train_data = JLD.load("train_digits.jld")["train_data"]
digits = [0, 1]
X = hcat(
    train_data[:, :, digits[1] + 1],
    train_data[:, :, digits[2] + 1]
)
X = float(X[:, 1:5:end]);
```

The [`Distances.jl`](https://github.com/JuliaStats/Distances.jl) package provides efficient code for computing pairwise point distances faster than our implementation.

```julia
GR.inline("svg")
D_digits = pairwise(Euclidean(), X, dims = 2);
Xdigit = dist2coords(D_digits,2)[1]
X0 = Xdigit[1:160, :]
X1 = Xdigit[161:320, :]
pdigits_truth_euclidean = scatter_nd(X0, label="$(digits[1])")
scatter_nd!(X1, label="$(digits[2])")
title!("Ground truth labels")
```

```julia
k = 2
@manipulate for iters in (0, 1, 10, 100)
    cluster_idx = simple_kmeans(Array(Xdigit'), k, iters)[1][:]
    pmds_digits_euclidean = scatter_nd(Xdigit; color=cluster_idx)
    title!("k-means after $(iters) iterations")
    plot(pmds_digits_euclidean, pdigits_truth_euclidean; layout=(1, 2), size=(700, 350))
end
```

```julia
D_digits = pairwise(CosineDist(), X, dims = 2)
Xdigit = dist2coords(D_digits, 2)[1]
X0 = Xdigit[1:160, :]
X1 = Xdigit[161:320, :]
pdigits_truth_cosine = scatter_nd(X0, label="$(digits[1])")
scatter_nd!(X1, label="$(digits[2])")
title!("Ground truth labels")
```

```julia
k = 2 
@manipulate for iters in (0, 1, 10, 100)
    cluster_idx = simple_kmeans(Array(Xdigit'), k,iters)[1][:]
    pmds_digits_cosine = scatter_nd(Xdigit; color=cluster_idx)
    title!("k-means: Cosine distance")
    plot(pmds_digits_cosine, pdigits_truth_cosine; layout=(1, 2), size=(700, 350))
end
```
